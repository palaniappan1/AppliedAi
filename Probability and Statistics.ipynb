# Random Number Generator (Uniform Distribution)

```python
import random
import pdb
print(random.random())
```

```python
# load IRIS dataset with 150 points
from sklearn import datasets
iris = datasets.load_iris()
d = iris.data
d.shape
```

```python
# Sample 30 points randomly from the 150 point dataset. This process is called uniform sampling.If we need to sample
# 30 points and call it dataset D' from dataset D of 150 points every point in D should be given equal chance to be a part of D' dataset. 
n = 150
m = 30
p = m/n
sampled_data = []
for i in range(0,n):
    if random.random() <= p:
#         pdb.set_trace()
        sampled_data.append(d[i,:])
len(sampled_data)
```

# Empirical Bootstrap Confidence Interval

```python
import numpy
from pandas import read_csv
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from matplotlib import pyplot

#load dataset
x = numpy.array([180,162,158,172,168,150,171,183,165,176])

# configure boootstrap
n_iterations = 1000
n_size = int(len(x))

medians = list()

for i in range(n_iterations):
    # prepare train and test sets
    s = resample(x, n_samples = n_size)
    m = numpy.median(s)
    medians.append(m)
    
pyplot.hist(medians)
pyplot.show()

#confidence intervals
alpha = 0.95
# (1- C) / 2 formula
p = ((1 - alpha) / 2.0) * 100 
lower = numpy.percentile(medians,p)

p = (alpha + ((1 - alpha) / 2.0)) * 100
upper = numpy.percentile(medians,p)

print("%.1f confidence interval %.1f and %.1f" % (alpha * 100, lower, upper))

```

# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

#generate a gaussian r.v X
x = stats.norm.rvs(size = 1000)
# array of size 1000 (Random variables) all these values are gaussian distributed as we use norm 
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw=0.5)
plt.show()
```

```python
stats.kstest(x,'norm')
# giving variable x and compairng it with normal distribution
# p-value is quite high so the R.V is normally distributed
```

```python
y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y),bw = 0.1)
plt.show()
```

```python
stats.kstest(y,'norm')
# p-value is 0 which means it is not a normally distributed R.V
```
